```python
from transformers import AutoTokenizer

old_tokenizer = AutoTokenizer.from_pretrained("gpt2")
```

```python
example = '''def add_numbers(a, b):
    """Add the two numbers `a` and `b`."""
    return a + b'''

tokens = old_tokenizer.tokenize(example)
tokens
```

```python
len(tokens)
```

```python
old_tokenizer.save_pretrained("code-search-net-tokenizer")
```

```python
bert_tokenizer = AutoTokenizer.from_pretrained('bert-base-cased')
```

```python
tokens = bert_tokenizer.tokenize(example)
tokens
```

```python
len(tokens)
```

```python
bert_tokenizer(example)['input_ids']
```

```python
bert_tokenizer(example)
```

```python
from transformers import pipeline

question_answerer = pipeline("question-answering")
context = """
Alice is sitting on the bench. Bob is sitting next to her.
"""
question = "Who is the CEO?"
question_answerer(question=question, context=context)
```
