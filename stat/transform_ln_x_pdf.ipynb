```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample_size = 100000
```

```python
x1 = np.random.uniform(0,1,sample_size)
x2 = np.random.uniform(0,1,sample_size)
```

```python
plt.hist(x1,density=True)
```

$$X \sim Uniform(0,1)$$
$$Y=ln(X)$$
$$pdf(X)=f_X(x)=1 \quad  for \  0 \leq x \leq 1$$
$$f_Y(y)=\frac{f_X(x)}{dy/dx}=\frac{f_X(e^y)}{1/x}=e^yf_X(e^y)=e^y$$

```python
y1 = np.log(x1)
count, bins, _ = plt.hist(y1, density=True)
def y1_pdf(y): return np.exp(y)
plt.plot(bins,y1_pdf(bins))
```

$$V=ln(Y)$$
$$f_V(v)=\frac{f_y(ln(v))}{v}$$

```python
v1=np.exp(y1)
count, bins, _ = plt.hist(v1, density=True)
def v1_pdf(v): return y1_pdf(np.log(v))/v
plt.plot(bins,v1_pdf(bins))
```

```python
y2 = x1*x2
plt.hist(y2,density=True)
```

```python
y4=np.exp(np.log(x1)+np.log(x2))
plt.hist(y4,density=True)
```

```python
y5 = np.log(x1)+np.log(x2)
count, bins, _ = plt.hist(y5, density=True)
def y2_log(y): return np.convolve(np.exp(y), np.exp(y))[::2]
plt.plot(bins,y2_log(bins))
# plt.plot(np.linspace(-10,10),y2_log(np.linspace(-10,10)))
```

```python
def y2_pdf(y):
    # Half-down sampling
    y2_log = lambda y: np.convolve(y1_pdf(y),y1_pdf(y))[::2]
    return y2_log(np.log(y))/y
```

```python
count, bins, _ = plt.hist(y2, density=True)
plt.plot(bins,y2_pdf(bins))
```
